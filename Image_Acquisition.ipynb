```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('orange.png')
cv2_imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
img_resize = cv2.resize(img_gray, (700, 500))
cv2_imshow(img_resize)
```

```python
img_flip = cv2.flip(img_resize, 0)
cv2_imshow(img_flip)
```
